```python
#CALCULATION OF TTR
import nltk
from operator import itemgetter
import string
import re
from nltk.tokenize import sent_tokenize, word_tokenize



file = open(r"Shakespere.txt", encoding="utf8")
text = file.read()
tokens = 0
wordcount = {}

#regexs to remove puntuations
reg = "[A-Za-z0-9]*"

#tokeniztion of words and calculate types and tokens
words = word_tokenize(text)
for word in words:
    word = word.lower()
    word = (re.match(reg,word)).group(0)
    if word =="":
        continue;
    tokens +=1
    if word not in wordcount:
        wordcount[word] = 1
    else:
        wordcount[word] += 1
        
#calculation of TTR
print("no. of tokens :=", tokens)
type = len(wordcount.keys())
print("type = :",type)
ttr = type/tokens
print("ttr  := ", ttr)
    
```

```python
print("                           ZIPF'S LAW      ")
print(" ->it states that frequrncy of words is inversly proprtion to rank")
print("   i.e 'f ‚àù 1/r'")

from matplotlib import pyplot
keys = []
length_words=[]
frequency =[]
#no of meaning find out using oxford dictionary
no_of_meanings = { "enough":9,"bark":2,"nail":4,"about":12,"window":5,"project":8}
for key, value in reversed(sorted(wordcount.items(), key = itemgetter(1))):
    keys.append(key)
    a = (len(key))
    length_words.append(a)
    #print(key,"   ",len(key))
    frequency.append(value)
    #print (key, value)
rank = [i for i in range(1,type+1)]

print("")

# plot graph of frequency and rank
pyplot.plot(rank,frequency)
pyplot.xlabel('rank')
pyplot.ylabel('frequency')
pyplot.title("Zipf's Law ")
pyplot.show()
print("#from above graph we find Ziph's LAw holds as f ‚àù 1/r")
print("---------------------")

#plot graph of length and frequency
pyplot.plot(frequency,length_words,'o')
pyplot.title("length of words VS frequency of words")
pyplot.xlabel('frequency')
pyplot.ylabel('length_words ')
pyplot.show()
print("#By comparing the curve of avove graph with graph of inversly proportion")
print("# we find word_length ‚àù 1/frequency")
print("---------------------")

# plot graph between word_length and ‚àö(frequency)
word_meanings = []
word_frequency = []
for key, value in no_of_meanings.items():
    if key in wordcount:
        word_frequency.append((wordcount[key])**(.5))
        word_meanings.append(value)

print("         ZIFH'S LAW FOR 'MEANING AND LENGTH'")
pyplot.title("meanings of words VS ‚àö(frequency of words)")
pyplot.xlabel('‚àö(word_frequency)')
pyplot.ylabel('word_meanings ')
pyplot.plot(word_frequency,word_meanings,'o')
pyplot.show()
print("# from above graph curve we find that find word_meanings ‚àù ‚àö(frequency of words)")
print("  =>ZIPF'S LAW HOLD TRUE FOR MEANING AND VALUE")
```

```python
#HEAP'S LAW
    
import numpy as np
paragraphs = text.split("\n\n")
segment_tokensCount = []
segment_wordCount = []
segment_tokens = 0
segment_word = {}

#here each paragraph is considered as a segment
#calculates the volabulary and tokens of segments
for segment in paragraphs:
    words = word_tokenize(segment.lower())
    if len(words) < 10:
        continue;
    for word in words:
        word = (re.match(reg,word)).group()
        if word =="":
            continue;
        segment_tokens +=1
        segment_tokens +=1
        if word not in segment_word:
            segment_word[word] = 1
        else:
            segment_word[word] += 1
    if segment_tokens != 0 and segment_word != 0:
        segment_tokensCount.append(segment_tokens)
        segment_wordCount.append(len(segment_word))

#print(segment_tokensCount)
#print(segment_wordCount)
pyplot.plot(segment_tokensCount,segment_wordCount,'o')
pyplot.xlabel('segment_tokensCount')
pyplot.ylabel('segment_wordCount')
pyplot.title("Heaps‚Äô law")
pyplot.show()

#calculate value of k and beta using slope and intercept of graph
print("value of k and beta using slope and intercept of graph")
slope, intercept = np.polyfit(np.log(segment_tokensCount), np.log(segment_wordCount), 1)
print("value of beta:", slope,"   value of k:", np.exp(intercept))

```
